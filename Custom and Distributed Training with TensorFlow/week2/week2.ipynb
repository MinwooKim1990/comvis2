```python
def fizzbuzz(x):
    for i in range(x):
        if (i+1) %15 == 0:
            print('fizz_buzz')
        elif (i+1) %3 ==0:
            print('fizz')
        elif (i+1) % 5 == 0:
            print('buzz')
        elif (i+1) %15 == 0:
            print('fizz_buzz')
        else:
            print(i+1)
    return 

fizzbuzz(100)
```
